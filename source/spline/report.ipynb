```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import CubicSpline
```

```python
error = []
x = np.linspace(0, 10, 1000)
y = np.exp(x)
with open("data5natural.txt") as f:
    y_old = [float(i) for i in f.readlines()]
    error.append(max(abs(y_old - y)))
with open("data10natural.txt") as f:
    y_old = [float(i) for i in f.readlines()]
    error.append(max(abs(y_old - y)))
with open("data20natural.txt") as f:
    y_old = [float(i) for i in f.readlines()]
    error.append(max(abs(y_old - y)))
with open("data40natural.txt") as f:
    y_old = [float(i) for i in f.readlines()]
    error.append(max(abs(y_old - y)))
with open("data80natural.txt") as f:
    y_old = [float(i) for i in f.readlines()]
    error.append(max(abs(y_old - y)))
with open("data160natural.txt") as f:
    y_old = [float(i) for i in f.readlines()]
    error.append(max(abs(y_old - y)))  
N = [5, 10, 20, 40, 80, 160]

plt.scatter(np.log10(N), np.log10(error), label='Данные')
plt.grid()
plt.xlabel('Логарифм числа узлов')
plt.ylabel('Логарифм ошибки')
k = np.polyfit(np.log10(N), np.log10(error), 1)
N_plot = np.log10(np.linspace(5, 160, 1000))
y_plot = np.polyval(k, N_plot)
plt.plot(N_plot, y_plot, color="red", zorder=0, label='Аппроксимация')
plt.legend()
k
```
