```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.interpolate import CubicSpline
```

$ N = 5$

```python
error = []
x = np.linspace(0, 10, 1000)
y = np.exp(x)
with open("data5.txt") as f:
    y_old = [float(i) for i in f.readlines()]
    error.append(max(abs(y_old - y)))
with open("data10.txt") as f:
    y_old = [float(i) for i in f.readlines()]
    error.append(max(abs(y_old - y)))
with open("data20.txt") as f:
    y_old = [float(i) for i in f.readlines()]
    error.append(max(abs(y_old - y)))
with open("data40.txt") as f:
    y_old = [float(i) for i in f.readlines()]
    error.append(max(abs(y_old - y)))
with open("data80.txt") as f:
    y_old = [float(i) for i in f.readlines()]
    error.append(max(abs(y_old - y)))
with open("data160.txt") as f:
    y_old = [float(i) for i in f.readlines()]
    error.append(max(abs(y_old - y)))  
N = [5, 10, 20, 40, 80, 160]

plt.scatter(np.log10(N), np.log10(error), label='Данные')
plt.grid()
plt.xlabel('Логарифм числа узлов')
plt.ylabel('Логарифм ошибки')
k = np.polyfit(np.log10(N), np.log10(error), 1)
N_plot = np.log10(np.linspace(5, 160, 1000))
y_plot = np.polyval(k, N_plot)
plt.plot(N_plot, y_plot, color="red", zorder=0, label='Аппроксимация')
plt.legend()
k
```

```python
error = []
ABOBA = [5, 10, 20, 40, 80, 160]
x = np.linspace(0, 2*np.pi, 1000)
y = np.sin(x)
for i in ABOBA:
    x_old = np.linspace(0, 2*np.pi, i)
    y_old = np.sin(x_old)
    cs = CubicSpline(x_old, y_old, bc_type='natural')
    error.append(max(abs(cs(x) - y)))

plt.plot(np.log10(ABOBA), np.log10(error))
np.polyfit(np.log10(ABOBA), np.log10(error), 1)


```
