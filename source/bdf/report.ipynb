```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("data_cube.txt", "r") as f:
    data = f.read()
datas = data.split("\n\n")

x = np.linspace(-6, 1, 12)

errors_1 = []
for data in datas:
    amogus = np.array([[float(a.split()[0]), float(a.split()[1])] for a in data.split("\n")])
    errors_1.append(max(abs(np.power(amogus[:, 0], 4)/4 - amogus[:, 1])))
plt.scatter(x[:-1], np.log10(errors_1)[:-1])
plt.grid()
plt.xlabel("Логарифм шага")
plt.ylabel("Логарифм ошибки")
```

```python
with open("data_oscillator.txt", "r") as f:
    data = f.read()
datas = data.split("\n\n")

x = np.linspace(-6, 1, 12)

errors_1 = []
for data in datas:
    amogus = np.array([[float(a.split()[0]), float(a.split()[1])] for a in data.split("\n")])
    errors_1.append(max(abs(np.sin(amogus[:, 0]) - amogus[:, 1])))
plt.scatter(x, np.log10(errors_1))
plt.grid()
coef = np.polyfit(x[5:], np.log10(errors_1[5:]), 1)
y = np.polyval(coef, np.linspace(-3, 1, 100))
plt.plot(np.linspace(-3, 1, 100), y)
plt.xlabel("Логарифм шага")
plt.ylabel("Логарифм ошибки")
coef
```
