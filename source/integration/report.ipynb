```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
CORRECT_ANSWER = 1.839071529076452

with open("N3.txt") as f:
    y = np.array([float(i) for i in f.readlines()])
    y = np.log10(np.abs([i-CORRECT_ANSWER for i in y]))
x = np.linspace(-3, 1, 1000)
plt.plot(x, y)
plt.grid()
plt.xlabel("Логарифм длины подотрезка")
plt.ylabel("Логариф ошибки")
np.polyfit(x[500:700], y[500:700], 1)
```
