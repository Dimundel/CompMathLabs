# Интерполяция

```python
import matplotlib.pyplot as plt
import numpy as np
NUM_OF_DOTS = 1000;
NUM_OF_LENGTHS = 1000;
```

## Зависимость логарфма ошибки интерполяции от логарифма длины отрезка
### N = 3 

```python
x = np.linspace(-7, 1, 1000)[1:]
y_error = []
with open("data3.txt", 'r') as f:
    for i in x:
        x_correct = np.linspace(0, 10**i, 1000)
        y_correct = np.exp(x_correct)
        y_old = [float(next(f)) for _ in range(1000)]
        y_error.append(max(abs(y_old - y_correct)))

y_chebishev = []
with open("data3chebishev.txt", 'r') as f:
    for i in x:
        x_correct = np.linspace(0, 10**i, 1000)
        y_correct = np.exp(x_correct)
        y_old_chebishev = [float(next(f)) for _ in range(1000)]
        y_chebishev.append(max(abs(y_old_chebishev - y_correct)))

plt.plot(x, np.log10(y_error), label='Равномерно')
plt.plot(x, np.log10(y_chebishev), color='r', label='Узлы Чебышева')

plt.grid()
plt.legend()
plt.xlabel("Логарифм длины отрезка")
plt.ylabel("Логарифм ошибки интерполяции")
np.polyfit(x[600:800], np.log10(y_error)[600:800], 1)
```
