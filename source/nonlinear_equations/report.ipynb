```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(1, 101, 1)

with open("e01.txt", 'r') as f:
    y1 = [float(i.split()[1]) for i in f.readlines()]
err1 = [abs(i - 0.86126488486817543718444234197066) for i in y1]
err1 = np.log10(err1)

with open("e02.txt", 'r') as f:
    y2 = [float(i.split()[1]) for i in f.readlines()]
err2 = np.array([abs(i - 0.94782822379959028291373723856607) for i in y2])
err2 = np.where(err2 != 0.0, np.log10(err2), np.log(0.000000000000000000001))

with open("e05.txt", 'r') as f:
    y3 = [float(i.split()[1]) for i in f.readlines()]
err3 = [abs(i - 1.2617030552531017713083444630671) for i in y3]
err3 = np.log10(err3)

with open("e08.txt", 'r') as f:
    y4 = [float(i.split()[1]) for i in f.readlines()]
err4 = [abs(i - 1.5853138613542082165987172986665) for i in y4]
err4 = np.log10(err4)

plt.plot(x, err1, label="e=0.1")
plt.plot(x, err2, label="e=0.2")
plt.plot(x, err3, label="e=0.5")
plt.plot(x, err4, label="e=0.8")
plt.grid()
plt.legend()
plt.xlabel("Номер итерации")
plt.ylabel("Логарифм ошибки")
plt.xlim(left=1, right=10)
```
