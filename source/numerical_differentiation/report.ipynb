## №2


1) $N = 3$ и {$x_0 - h, x_0, x_0 + h, x_0 + 2h$}

```python
x3 = []
y3 = []
with open('dataN3.txt', 'r') as f:
    data = f.readlines()
for row in data:
    x3.append(float(row.split()[0]))
    y3.append(float(row.split()[1]))
plt.plot(x3, y3, label='N=3')
plt.grid()
plt.xlabel('Логарифм шага')
plt.ylabel('Логарифм ошибки')
plt.legend()
```
